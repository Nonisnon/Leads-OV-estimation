```python
import pandas as pd
import h2o
from h2o.automl import H2OAutoML
```

```python
train = pd.read_csv("../data/processed/train.csv")
train
```

```python
pred = pd.read_csv("../data/processed/predict.csv")
pred
```

```python
h2o.init()
```

```python
h2train = h2o.H2OFrame(train)
h2pred = h2o.H2OFrame(pred)
```

```python
automl1 = H2OAutoML(seed=42, max_runtime_secs=1200, sort_metric='RMSE')
automl1.train(x, y, training_frame=h2train)
```
