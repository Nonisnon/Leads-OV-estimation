## MERGE DFs

```python
tadata=pd.read_csv('../data/processed/tadatageo.csv')
tadata= pd.DataFrame(tadata)
tadata
```

```python
train = pd.merge(tadata,  
                roodata,  
                on ='Latitude',  
                how ='left') 
train
```

```python
train = train[['name', 'priceRange', 'rating', 'reviewCount', 'postCode', 'Latitude', 'Longitude_x','Weekly orders']]
train['value'] = train['rating']*train['reviewCount']
train = train.drop_duplicates(subset=['name'])
predict = train[train['Weekly orders'].isna()]
predict = predict.drop(columns=['Weekly orders'])
predict = predict.reset_index(drop=True)
```

```python
predict.to_csv(f'/Users/Nonis/desktop/Leads-OV-estimation/data/processed/predict.csv', index=False)
predict
```
